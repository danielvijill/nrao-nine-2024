### Accesing to the NRAO network

In order to access the NRAO network we need to introduce the command ```ssh username@ssh.cv.nrao.edu```, this will take you to polaris, do not run any heavy commands like casa in here.

If we wish to process data is is necessary that we work on a cvpost node we reserved, in order to reserve a node for, i.e. 5 days, we can type in the console ```ssh cvpost-master```, and then you can write ```nodescheduler -r 5```, this will reserve 1 node for 5 days.

Once your node has been assigned, you will receive an email indicating the node you have been assigned or you can check by writing ```squeue``` in the cvpost-master.

Next you move into your cvpost with the command ```ssh cvpost###``` (introduce your assigned node number)

### Moving through file systems

In your console you can use basic linux commands to make your way through folders in your local computer and the lustre file system

```pwd``` will display the current path you are in

```ls``` will display the files and folders in your current location

```ls -l``` will display the files and folders in your current location in the form of a list

```ls -a``` will display the files and folders in your current location including the hidden ones

```cd <foldername>``` will allow you to move to a specific folder

```cd ..``` will allow you to move to the previous folder

```mkdir <foldername>``` will create a new folder

```rm -rf``` will forcefully and recursively remove a file or folder

### Downloading Data

In order to download data you should move to your ```DATA``` folder, which is in lustre, then you create a folder for the observations of this project, let's say we name it ```V960 Mon``` since it is the name of the object example we will work with.

We need to head over the [ALMA archive](https://https://almascience.nrao.edu/aq/) in which we will be able to download the raw data from ALMA observations by browsing the project code, in this case ```2016.1.00209.S```, corresponding to a low-mass star formation survey directed by Michihiro Takami.

image.png

Once we have the project observations listed, we head over the first V960 Mon observation, which was taken with the 7m configuration of the ALMA antennas, then we hit the icon on the left corresponding to "explore and download". For our purposes we then select the option "Open legacy Request Handler" in which we can select only the raw data and not the products.

image.png

We make sure to check only the "readme", "auxiliary" and "raw" options. The you click on "download selected", this should initiate a download of a ```.sh``` file.

This file should then be moved to the folder in which you wish to download your data. One recommendation would be to create a folder for the 7m array observations since then we will be processing the 12m array observations.

Once the download file is in the destination folder you need to make the file executable by typing ```chmod u+x download*.sh```, then you execute the file with the command ```.download*.sh``` and the download should start.

### Starting CASA

In order to start CASA in the version we want and the parameters we desire, we head to the folder in which we wish to run casa and then introduce the command line ```casa --pipeline -r 4.7.2-el7 --nologger``` in the console. This will ensure that we log into the pipeline version of casa in the release 4.7.2, since the data we will work with can only be reduced with versions corresponding to the time of their observations.
